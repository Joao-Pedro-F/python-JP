```python
def dois_soma():
    n1 = int(input('Digite um número: '))
    n2 = int(input('Digite outro número: '))
    soma = n1+n2
    print(soma)


def dois_sub():
    n1 = int(input('Digite um número: '))
    n2 = int(input('Digite outro número: '))
    sub = n1-n2
    print(sub)


def dois_mult():
    n1 = int(input('Digite um número: '))
    n2 = int(input('Digite outro número: '))
    mult = n1*n2
    print(mult)


def dois_divi():
    n1 = int(input('Digite um número: '))
    n2 = int(input('Digite outro número: '))
    div = n1/n2
    print(div)


def dois_exp():
    n1 = int(input('Digite um número: '))
    n2 = int(input('Digite outro número: '))
    exp = n1**n2
    print(exp)


def menu():
    while True:
        print('='*25)
        print('Calculadora')
        print('='*25)
        print()
        print(' [1] - Somar')
        print(' [2] - Subtração')
        print(' [3] - Multiplicação')
        print(' [4] - Divisão')
        print(' [5] - Exponenciação')

        esc = int(input('Digite uma opção: '))
        if esc == 1:
            dois_soma()
        elif esc == 2:
            dois_sub()
        elif esc == 3:
            dois_mult()
        elif esc == 4:
            dois_divi()
        elif esc == 5:
            dois_exp()
        else:
            print('Tente Novamente')

        x = str(input('Continuar? (Sim/Não): '))
        if x == 'Sim':
           continue
        elif x == 'Não3':
           break
        else:
            print('Fim da Execução')






menu()
```
