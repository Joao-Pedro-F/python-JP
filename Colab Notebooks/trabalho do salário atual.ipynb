```python
salario_atual = input("Qual é o salário atual do funcionário? ")
percentual = input("Qual é o percentual de aumento desejado? ")

if salario_atual and percentual:  # Verifica se os valores não são vazios
    salario_atual = float(salario_atual)  # Converte o salário atual para um número
    percentual = float(percentual)  # Converte o percentual para um número

    acrescimo = salario_atual + (salario_atual * percentual / 100)  # Calcula o novo salário após o aumento

    print(f"O seu salário no mês que vem será de {acrescimo} R$")
else:
    print("Insira valores válidos")  # Se o usuário não inserir valores válidos, exibe uma mensagem
```
